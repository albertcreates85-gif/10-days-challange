```python
#our data is using 17.5+ MB so we should reduce it .
dtypes_mapping = {
    'popularity': 'int8',
    'duration_ms': 'int8',
    'energy': 'int8',
    'time_signature': 'int8',
    'key': 'int8',
    'track_genre': 'category',
    'tempo': 'int16',
    'mode': 'int8',
    'loudness': 'int8',
    'danceability': 'int8',
    'speechiness':'int8',
    'acousticness':'int8',
    'liveness':'int8',
    'instrumentalness':'int8',
    'valence':'int8'
    }

for col, dtype in dtypes_mapping.items():
    df[col] = df[col].astype(dtype)
```

```python
#you can see our memory consuption reduced massively from 17.5 mb to 5.3 mb which is huge almost 75% reduced
```

```python
#top 10 artists
sns.set_style("white")
top_artist = df['artists'].value_counts().nlargest(10)
sns.barplot(y=top_artist.index , x=top_artist.values,palette="deep" )
plt.title('top 10 artists')
```

```python
top_artist = df['artists'].value_counts().nlargest(10)
top_pop = df['popularity'].nlargest(10)
sns.barplot(y=top_artist.index , x=top_pop ,palette="deep" )
plt.title('top 10 artists by their popularity')
```

```python
plt.figure(figsize=(10, 5))
top_albums = df['album_name'].value_counts().nlargest(10)
sns.barplot(y=top_albums.index, x=top_albums.values , palette="pastel")
plt.title("Top 10 Albums")
plt.xlabel("Count")
plt.ylabel("Album Name")
plt.show()
top_albums
```

```python
plt.figure(figsize=(10, 5))
top_artists = df['track_name'].value_counts().nlargest(10)
sns.barplot(y=top_artists.index, x=top_artists.values ,palette='deep')
plt.title("Top 10 Track Name")
plt.xlabel("Count")
plt.ylabel("Track Name")
plt.show()
top_artists
```

```python
avg_popularity_by_genre = df.groupby('track_genre')['popularity'].mean().reset_index()
top10_popular_genres = avg_popularity_by_genre.nlargest(10, 'popularity')

top10_popular_genres
#if you figure out how to plot graph here do let me know
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
